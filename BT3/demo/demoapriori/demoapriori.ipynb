```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
store_data = pd.read_csv("store_data.csv", header=None) #Đọc file tải lên
display(store_data.head())
print(store_data.shape)
```

```python
records = [] #Chuẩn hóa csdl gốc cho đúng dạng tham số  truyền vào hàm apriori
for i in range(1, 7501):
    records.append([str(store_data.values[i, j]) for j in range(0, 20)])
```

```python
print(records)
```

```python
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2) #Chạy thuật toán và lưu kết quả
association_results = list(association_rules)
```

```python
for i in range(0, len(association_results)):#Tập phổ biến phù hợp với yêu cầu
    print(association_results[i][0])
```

```python
for item in association_results: #In ra các luật kết hợp kết quả cùng các chỉ số đánh giá của nó
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
