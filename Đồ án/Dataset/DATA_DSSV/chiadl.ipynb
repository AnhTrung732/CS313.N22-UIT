## XỬ LÍ

```python
# Xóa khoảng trắng ở trước tên môn
monhoc['mamh'] = monhoc['mamh'].str.replace(' ', '')

#CS1113 = CS113
monhoc['mamh'] = monhoc['mamh'].replace({'CS1113': 'CS113'})
```

```python
print(monhoc['mamh'].value_counts())
monhoc['mamh'].unique()
```

```python
monhoc['mamh'] = monhoc['mamh'].astype(str)  # Chuyển đổi kiểu dữ liệu thành chuỗi
# tách độ khó
monhoc['tenmon'] = monhoc['mamh'].apply(lambda x: re.split('\d+', x)[0].strip())
monhoc['dokho'] = monhoc['mamh'].apply(lambda x: int(re.search('(\d+)', x).group(1)[:1]) if re.search('(\d+)', x) else '')
```

```python
# định nghĩa makhoa: 
# 0 - môn chung
# 1 - KHMT
# 2 - CNPM 
# 3 - HTTT
# 4 - MMT
# 5 - KTMT
# 6 - KTTT

#LƯU Ý:
# Sự phân chia trên có thể chưa chính xác hoàn toán
# Có các môn độc lập hoặc tương đương nhau, nhưng được tách ra và được quản lí bởi các khoa khác nhau
# Có các môn học tuy giống nhau nhưng khác mã được sửa đổi và tích hợp từ các khóa cũ

# Các môn IT thuộc rất nhiều khoa quản lí
# vd: IT004 Hướng đối tượng IT005 mạng máy tính IT006 kiến trúc máy tính
# nhưng vì đều có tính chất là các môn nhập môn và là môn tiên quyết
# nên có thể xem như các môn chung

def encode_khoa(monhoc):
    if monhoc in ['VCPL','VCPH', 'SOCI','PHIL','PEDU', 'MLPE', 'MAT','LIA','JAN','ITEW', 'HCMT','ENG', 'ENGA', 'ENGBT', 'ENGL','ADENG','ITEM','IT', 'EN', 'ENBT', 'MA', 'PE',"SMET",'SS']:
        return 0
    elif monhoc in ['DSAL','CSKI', 'CSC','CSBU', 'CS']:
        return 1
    elif monhoc in ['OOPT', 'SE','WINP']:
        return 2
    elif monhoc in ['STA','SP', 'MKTG','IS','INT','IEM', 'ECON', 'DBSS','ENGL','ADENG','EC','IS','MATH','MSIS','PHYS','SC','SPCH','STAT','ACCT']:
        return 3
    elif monhoc in ['OSYS','CNET','CNBU','NT','MAH']:
        return 4
    elif monhoc in ['CARC','PH','CE']:
        return 5
    elif monhoc in ['DS','CU','IE']:
        return 6
    else:
        return None
    
#tạo thuộc tính makhoa
monhoc['makhoa'] = monhoc['tenmon'].apply(encode_khoa)
```

```python
monhoc = monhoc[~monhoc['trangthai'].isin([0, 4, 5])]
monhoc['trangthai'].unique()
```

```python
monhoc = monhoc[monhoc['dokho'] != '']
monhoc['dokho'].unique()
```

```python
# Kiểm tra tính đồng bộ của cột "sotc" trong mỗi môn học

inconsistent_mh = monhoc.groupby('mamh')['sotc'].nunique().loc[lambda x: x > 1].index.tolist()
if len(inconsistent_mh) > 0:
    print("Có môn học không đồng bộ về sotc:", inconsistent_mh)
    print()

   # Tạo bảng đếm giá trị sotc theo môn học
    sotc_counts = monhoc.groupby('mamh')['sotc'].value_counts().reset_index(name='count')

    # Lặp qua từng môn học không đồng bộ
    for mamh in inconsistent_mh:
        sotc_values = sotc_counts.loc[sotc_counts['mamh'] == mamh, 'sotc'].tolist()
        sotc_counts_grouped = sotc_counts.loc[sotc_counts['mamh'] == mamh, ['sotc', 'count']].groupby('sotc').sum()
        unique_sotc = sotc_counts_grouped.index.tolist()
        count_values = sotc_counts_grouped['count'].tolist()

        # Hiển thị thông tin số lượng dữ liệu cho từng giá trị sotc
        print(f"Môn học {mamh} có các giá trị sotc khác nhau và số lượng dữ liệu tương ứng:")
        for i in range(len(unique_sotc)):
            print(f"- Giá trị sotc {unique_sotc[i]}: {count_values[i]} dữ liệu")
else:
    print("Tất cả các môn học đều đồng bộ về sotc.")
```

```python
# Tạo bảng đếm giá trị sotc theo môn học
sotc_counts = monhoc.groupby('mamh')['sotc'].value_counts().reset_index(name='count')

# Lặp qua từng môn học
for mamh in monhoc['mamh'].unique():
    # Lấy giá trị 'sotc' có số lượng dữ liệu lớn nhất cho môn học hiện tại
    max_sotc = sotc_counts.loc[sotc_counts['mamh'] == mamh, 'sotc'].iloc[0]
    # Đồng bộ giá trị 'sotc' của môn học về giá trị max
    monhoc.loc[monhoc['mamh'] == mamh, 'sotc'] = max_sotc
```

```python
# Tách dữ liệu từ trước 2022 và từ 2022 trở đi
monhoc_test = monhoc[monhoc['namhoc'] >= 2022]
monhoc_train = monhoc[monhoc['namhoc'] < 2022]
```

```python
# Nhóm dữ liệu theo mssv và thu thập thông tin môn học
grouped_data = monhoc_test.groupby('mssv').agg({
    'mamh': lambda x: x.tolist(),
    'makhoa': lambda x: x.tolist(),
    'sotc': lambda x: x.tolist(),
    'dokho': lambda x: x.tolist(),
    'trangthai': lambda x: x.tolist()
})
```

```python
# Tạo danh sách các sinh viên với thông tin môn học
sinhvien_list = []
for mssv, row in grouped_data.iterrows():
    sinhvien = {
        'mssv': mssv,
        #thong tin mon hoc
        'mamh': row['mamh'],
        'makhoa': row['makhoa'],
        'sotc': row['sotc'],
        'dokho': row['dokho'],
        'trangthai': row['trangthai'],

        #thong tin hoc phan cua sv
        'namhoc': monhoc_test.loc[monhoc_test['mssv'] == mssv, 'namhoc'].tolist(),
        'hocky': monhoc_test.loc[monhoc_test['mssv'] == mssv, 'hocky'].tolist(),
        'diemtb': monhoc_test.loc[monhoc_test['mssv'] == mssv, 'diem_hp'].tolist()
    }
    sinhvien_list.append(sinhvien)


```

```python
with open('dssv_test.txt', 'w') as file:
    for i, sinhvien in enumerate(sinhvien_list, start=1):
        file.write("Sinh_vien_{}: \n".format(i))
        file.write("MSSV: " + str(sinhvien['mssv']) + "\n")
        file.write("monhoc: " + str(sinhvien['mamh']) + "\n")
        khoa_nguyen = [int(value) if not math.isnan(value) else 0 for value in sinhvien['makhoa']]
        file.write("khoa: " + str(khoa_nguyen) + "\n")

        file.write("sotc: " + str(sinhvien['sotc']) + "\n")
        file.write("dokho: " + str(sinhvien['dokho']) + "\n")
        file.write("trangthai: " + str(sinhvien['trangthai']) + "\n")
        file.write("namhoc: " + str(sinhvien['namhoc']) + "\n")
        file.write("hocky: " + str(sinhvien['hocky']) + "\n")
        file.write("diemtb: " + str(sinhvien['diemtb']) + "\n")
        file.write("\n")
```

```python
# Chuyển đổi danh sách sinh viên thành DataFrame
df = pd.DataFrame(sinhvien_list)

# Ghi dữ liệu vào tệp CSV
df.to_csv('dssv_test.csv', index=False)
```
